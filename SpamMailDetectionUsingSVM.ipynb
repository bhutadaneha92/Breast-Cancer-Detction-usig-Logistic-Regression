```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer #Extract the feature from text
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score
import chardet
```

# Data Preprocessing

```python
#To determine character encoding

with open('spam.csv', 'rb') as f:
    result = chardet.detect(f.read(100000))
print(result)

#load the dataset
mail_data = pd.read_csv('spam.csv', encoding = result['encoding'])
print(mail_data)

#Drop irrelevant columns and rename columns
mail_data.drop(['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4'],axis=1, inplace=True) #axis=1 means columns drop
mail_data.rename(columns={'v1':'Target','v2':'Text'}, inplace=True)
print(mail_data)
print(mail_data.shape)

#replace all the null values with null string
#data = mail_data.where((pd.notnull(mail_data)), '')

```

```python
mail_data.head()
```

```python
#label define
# ham = 1, spam = 0
mail_data.loc[mail_data['Target'] == 'spam', 'Target'] = 0
mail_data.loc[mail_data['Target'] == 'ham', 'Target'] = 1
mail_data.head()
```

```python
#Seperate the data into text and labels. X-->text, Y-->Labels
X= mail_data['Text']
Y= mail_data['Target']
print(X,Y)
```

# Spliting of data

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, train_size=0.8,test_size=0.2,random_state=1)
print(X_train.shape,X_test.shape)
```

# Feature Extraction

```python
#transform the text data into feature vectors that can be used as an input to the SVM model using TfidfVectorizer
#Convert the text to lower case letters
feature_extraction = TfidfVectorizer(min_df=1, stop_words='english', lowercase=True)
X_train_features = feature_extraction.fit_transform(X_train)
X_test_features = feature_extraction.fit_transform(X_test)
print("X_train_features\n",X_train_features)

#Convert Y_train and Y_test data into integers
Y_train = Y_train.astype('int')
print("Y_train_int\n",Y_train)
Y_test = Y_test.astype('int')
```
